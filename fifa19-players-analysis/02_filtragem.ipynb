# FIFA 19 — Manipulação Básica

## 1. Carregando o dataset

```python
import pandas as pd

df = pd.read_csv('fifa.csv',index_col=0)
df.head()
```

## 2. Fatiamento

Criando um novo df com as características principais dos jogadores

```python
df = df[['Name','Age','Nationality','Club','Overall','Potential']]
df.head()
```

## 3. Filtro Simples

Filtragem simples do país mais vencedor da Copa do Mundo FIFA

```python
filtro_brasil = df.loc[df['Nationality'] == 'Brazil']
```

```python
filtro_brasil.describe()
```

## 4. Filtro Múltiplo

```python
filtro_barca = df.loc[df['Club'].str.contains('Barcelona', na=False)]
```

```python
filtro_barca.describe()
```

```python
filtro_brasil_over = df.loc[(df['Nationality'] == 'Brazil') & (df['Overall'] >= 85)]
filtro_brasil_over.sort_values('Overall', ascending=False)
```

```python
filtro_brasil_over.describe()
```

```python
filtro_pot = df.loc[df['Potential'] > 90]
filtro_pot.sort_values('Age', ascending=True)
```

```python
filtro_pot.describe()
```
