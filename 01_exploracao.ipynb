```python
import pandas as pd

df = pd.read_csv('fifa.csv', index_col=0)
df.head()
```

```python
#linhas e colunas
df.shape
```

```python
#tipos de dado e contagem de não-nulos por coluna
df.info()
```

## 3. Valores nulos

```python
#filtra apenas colunas com pelo menos 1 nulo
nulos = df.isnull().sum()
nulos[nulos > 0]
```

## 4. Valores únicos

```python
#quantidade de nacionalidades
df['Nationality'].nunique()
```

```python
#quantidade de clubes
df['Club'].nunique()
```

## 5.Overall

```python
df['Overall'].describe()
```

```python
#exibe nome, clube e o overall dos 5 melhores jogadores
df[['Name', 'Club', 'Overall']].sort_values('Overall', ascending=False).head(5)
```

## 6. Idade Media

```python
print(f'A idade média é de: {df['Age'].mean().round(1)}')
```

## Conclusões

* O dataset tem **18.207 jogadores** e **88 colunas de atributos**.
* `Release Clause` tem 1.564 nulos — jogadores sem multa rescisória definida em contrato.
* O Overall médio é **66.2**, com máximo de **94** (Messi e Cristiano Ronaldo).
* Há jogadores de **164 nacionalidades** em **651 clubes** diferentes.
