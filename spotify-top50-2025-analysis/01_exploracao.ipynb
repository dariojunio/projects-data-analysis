### Conclusões

* O dataset contém dados das 50 músicas mais ouvidas em 2025 na plataforma **Spotify**, contendo dados como bpm das músicas, ano de lançamento e até valores como dançabilidade, duração da música em segundos. o df é formado por **50** linhas e **15** colunas, com nenhum dado nulo/vazio

  Usando o comando describe, conseguimos informações interessantes:

  **Médias**

  bpm: **109.8** - as músicas seguem um padrão de moderado a animado,
  abrangendo todo tipo de gosto musical.

  duration_seconds: **205.44** (aproximadamente 3 min 25 s) - um valor
  considerável levando em conta o imediatismo que a sociedade vive hoje.

  danceability: **0.681** - assim como o bpm, está na média. São músicas
  que você pode dançar ou simplesmente curtir a vibe.

* O dataset revela um domínio claro dos **EUA**, com **36 dos 50 artistas (72%)**
  — o mercado global de streaming ainda é amplamente americano. O **Reino Unido**
  aparece em segundo com apenas 3 artistas, seguido de **Coreia do Sul** e
  **Colômbia** com 2 cada.

* Em relação aos gêneros, o **Pop** lidera com 10 músicas, mas se somarmos
  todas as suas variações — Indie Pop, Synth-Pop, K-Pop/Pop, Pop Rock —
  o domínio é ainda maior, representando mais da metade do ranking.

* **Sabrina Carpenter** é a artista mais presente, com **4 músicas** no top 50
  — única a ultrapassar 3 aparições. Billie Eilish, Chappell Roan e Taylor Swift
  aparecem com 3 músicas cada, domínio absurdo de cantoras do sexo feminino.
