# Spotify Top 50 Musics in 2025 - Outliers e Casos Fora do Padrão

### 1. Carregando o dataset

```python
import pandas as pd

#index_col=0 impede que o pandas crie uma coluna extra de índice
df = pd.read_csv('2025_top50_songs.csv', index_col=0)
df.head()
```

###2. Música mais antiga do ranking

```python
antigas = df.loc[df['release_year'] < 2000]
antigas[['song_title', 'artist', 'release_year', 'streams_2025_billions', 'peak_global_chart_position']]
```

### 3. Músicas com menos streams do ranking

```python
df.sort_values('streams_2025_billions',).head(5)
```

### 4. Caso Clairo

```python
clairo = df.loc[df['artist'] == 'Clairo']
clairo[['song_title', 'streams_2025_billions', 'peak_global_chart_position', 'danceability', 'energy', 'valence']]
```

### 5. Distribuição geral de streams

```python
df.sort_values('streams_2025_billions', ascending=False)[['song_title', 'artist', 'streams_2025_billions']]
```

### 6. Conclusões

* O caso mais surpreendente do ranking é **Kate Bush** com "Running Up That Hill",
  lançada em **1985** e presente no top 50 de 2025 com **0.62 bilhões** de streams.
  A música voltou às paradas globais após aparecer em Stranger Things em 2022, e novamente em 2025,
  provando que cultura pop e streaming podem ressuscitar músicas décadas depois
  do lançamento.

* A diferença entre o primeiro e o último lugar em streams é de apenas **1.14
  bilhões** (1.70 vs 0.56). Para um top 50 global, isso mostra que o piso de
  entrada é alto, qualquer música aqui já é um fenômeno por si só.

* **Clairo** é o outlier mais curioso da lista: posição **22** no chart global,
  a mais baixa do ranking inteiro, com apenas **0.63 bilhões** de streams e
  perfil sonoro bem diferente da média, baixa energia (0.52) e danceability
  (0.41). Ela chegou aqui não pelo impacto imediato, mas por uma base de fãs
  fiel que consome de forma consistente.

* O bottom 5 em streams tem nomes como **Shake It Off** (Taylor Swift) e
  **Espresso Remix** (Sabrina Carpenter), mostrando que até as "menores"
  músicas do ranking pertencem a artistas dominantes em 2025.
