# Spotify Top 50 Musics in 2025 - Volume vs Consistência

### 1. Carregando o dataset

```python
import pandas as pd
#index_col=0 impede que o pandas crie uma coluna extra de índice
df = pd.read_csv('2025_top50_songs.csv', index_col=0)
df.head()
```

### 2. Streams totais por artista

```python
#coloquei dentro de uma variavel para fazer a métrica musica/streams
artist_streams = df.groupby('artist')['streams_2025_billions'].sum().sort_values(ascending=False)
artist_streams
```

### 3. Músicas por artista

```python
#soma da quantidade de músicas por artista
artist_musics = df['artist'].value_counts()
artist_musics
```

### 4. Métrica de eficiência
Streams totais dividido pelo número de músicas - quem faz mais com menos.

```python
#métrica música/streams, utilizei ascending=False para mostrar do maior para o menor
metrics = artist_streams / artist_musics
metrics.sort_values(ascending=False)
```

## 5. Conclusões

* **Sabrina Carpenter** domina em volume absoluto com **5.24 bilhões** de streams,
  resultado direto da consistência com **4 músicas** no ranking. Nenhum outro
  artista chegou perto em streams totais.

* Porém, quando olhamos eficiência (streams por música) a história muda.
  **Lady Gaga & Bruno Mars** lideram com **1.70 bilhões por música**, contra
  **1.31** de Sabrina. Um único hit superou a média de qualquer artista consistente.

* A conclusão é que **as duas estratégias funcionam**, mas de formas diferentes:
  consistência garante volume total, mas um hit certeiro pode ser mais eficiente
  do que várias músicas medianas.
